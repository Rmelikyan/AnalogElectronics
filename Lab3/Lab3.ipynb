```python
import matplotlib.pyplot as plt
import numpy
```

```python
Vin = [.01, .015, .02, .025, .03, .035, .04, .045, .05, .055, .06, .065, .07, .075, .08]
Vout = [2.52, 3.72, 5.04, 6.24, 7.40, 8.42, 9.60, 10.6, 11.1, 12.1, 12.7, 13.0, 13.3, 13.5, 13.8]
Av = []
for i in range(len(Vout)):
    Av.append(Vout[i]/Vin[i])
```

```python
%matplotlib notebook
plt.figure(1)
plt.plot(Vin, Vout, '.-r')
```

```python
plt.figure(2)
plt.plot(Vin, Av, '.-g')
```
